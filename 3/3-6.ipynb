```python
import os
import pandas as pd
import numpy as np
base_url = '../../practical-jupyter/anime/'
```

```python
anime_csv = os.path.join(base_url, 'anime.csv')
df = pd.read_csv(anime_csv)

df.loc[df['episodes'] == 'Unknown'].head()
```

```python
df.where(df['rating'] < 9.2).head()
```

```python
df.loc[74, 'episodes'] = np.nan
df.loc[74, 'episodes']
```

```python
# episodesがUnknownの場合に　NaN に変更
df.loc[df['episodes'] == 'Unknown', 'episodes'] = np.nan
df
```

```python
# episodesがNaN(欠損値)の場合のみ抽出
df.loc[df['episodes'].isnull()].head()
```

```python
# 欠損値を持ったデータを除外,73,74がなくなる
# 非破壊的操作
df.dropna().loc[70:].head()
```

```python
# 破壊的操作
df.dropna(inplace=True)
df.loc[70:].head()
```

```python
# Seriesのデータ型確認
df['anime_id'].dtype
```

```python
# DataFrameのデータ型確認
df.dtypes
```

```python
# 型変換
pd.options.display.max_rows = 10 #出力制限
df['episodes'].astype(np.int64)
```

```python
df.astype({'episodes': np.int64, 'rating': np.float64})
```

```python
#astypeは非破壊的操作であることを確認
df.dtypes
```

```python
# 型を書き換える場合は代入する
df['episodes'] = df['episodes'].astype(np.int64)
df.dtypes
```

```python
df.sort_values('rating', ascending=False).head()
```

```python
import html

# mapメソッドを使うことで、Seriesの値ごとに関数を適用できる
print(df['name'].head())
print(df['name'].map(html.unescape).head())
```

```python
# applyメソッドではDataFrameの列または行に関数を適用
# 列の場合
df.apply(len)
```

```python
# 行の場合
df.apply(len, axis=1).head()
```

```python
df.apply(type)
```

```python
# applyメソッドにはラベルを指定
# name列とgenre列の文字数を合計
df.apply(lambda x: len(x['name']) + len(x['genre']), axis=1).head()
```

```python
# DataFrameの各値に対してlen()を適用
df[['name', 'genre']].applymap(len).head()
```

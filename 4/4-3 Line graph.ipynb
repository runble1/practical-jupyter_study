# 4-3 折れ線グラフ

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib notebook
%matplotlib inline
```

## 折れ線グラフを作成する

```python
'''
plot()メソッドにリスト型のデータを渡す
'''

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1, 3])

plt.show()
```

```python
'''
plot()メソッドにX値とY値を渡す
他にもリストラ行くオブジェクト（リスト・タプル・numpy.ndarray・pandas.Series）を渡せる
'''

fig = plt.figure()
ax = fig.add_subplot(111)

x = [0, 2, 4]
y = [0, 4, 2]

ax.plot(x, y)

plt.show()
```

```python
'''
複数の線を描画する
'''

fig = plt.figure()
ax = fig.add_subplot(111)

x = [0, 2, 4]
y1 = [0, 4, 2.5]
y2 = [4, 0, 1.5]

ax.plot(x, y1)
ax.plot(x, y2)

plt.show()
```

## 折れ線グラフを活用する

```python
import os
import pandas as pd
base_url = '../../practical-jupyter/anime/'
anime_stock_returns_csv = os.path.join(base_url, 'anime_stock_returns.csv')
df = pd.read_csv(anime_stock_returns_csv, index_col=0, parse_dates=['Date'])
df.head()
```

```python
'''
TOEI ANIMATIONおよび IG Port の株価騰落率を時系列で描画
'''

fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(111)

# データと判例を指定
ax.plot(df.index, df['TOEI ANIMATION'], label='TOEI ANIMATION')
ax.plot(df.index, df['IG Port'], label='IG Port')

# タイトル、軸ラベルを指定
ax.set_title('株価騰落率 2年間の推移')
ax.set_ylabel('株価騰落率')
ax.set_xlabel('年月')

# 凡例を有効化
ax.legend()

plt.show()
```

## 2軸を持つグラフを描画する

```python
'''
X軸を共有した2つのY軸を持つ図を作成する
twiny()メソッドを使用する
株価を折れ線グラフ、出来高は棒グラフで示す
'''

t4816_csv = os.path.join(base_url, '4816.csv')
df = pd.read(t4816_csv, index_col=0, parse_dates=['Date'])

fig = plt.figre(figsize=(10, 4))
ax1 = fig.add_subplot(111)

ax1.plot(df.index, df['Close'], color='b', label='株価')

# X軸を共有しY軸を2つ使用する設定
ax2 = ax1.twinx()
ax2.bar(df.index, df['Volume'], color='g', label='出来高', width=2)

# 軸と軸ラベルの設定
ax1.set_yticks([i * 2000 for i in range(5)])
ax1.set_ylabel('株価')
ax2.set_yticks([i * 50000 for i in range(5)])
ax2.set_ylabel('出来高')
ax1.set_xlabel('年月')

# グラフタイトルの設定
ax1.set_title('株価と出来高')
```
